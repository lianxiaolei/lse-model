```python
import sys
sys.path.insert(0,'../')
from collections import OrderedDict
import numpy as np
import numpy.linalg as llg
from scipy.optimize import minimize
import matplotlib.pylab as pl
from sklearn.model_selection import train_test_split
```

```python
import LSE
reload(LSE)
lse = LSE.LSEModel()
```

```python
# 真值
m = 100
n = 3
k = 1
s = np.random.normal(1.5, 0.5, (m, k))
s[s < 0.0001] = 0.0001
a = np.random.normal(1.5, 0.5, (k, n))
a[a < 0.0001] = 0.0001
gm_s = np.random.normal(0, .1, (m, 1))
gm_q = np.random.normal(0, .1, (1, n))
```

```python
u = np.zeros((m, n))
norm_a = llg.norm(a, axis=0).reshape(1, -1)
dots = np.dot(s, a)
delta = lse.vec_delta(dots, norm_a, gm_s, gm_q)
p = lse.prob(delta)
rand = np.random.random(p.shape)
u[rand < p] = 1.0
u[rand >= p] = 0.0
u.shape
```

```python
lse.fit(u, k, 10)
ss, aa, gms, gmq = (lse.est_s, lse.est_a, lse.est_gm_s, lse.est_gm_q)
```

```python
lse.likehood
```

```python
np.sum(u * np.log(p) + (1 - u) * np.log(1 - p))
```

```python
# with gamma s
norm_aa = np.linalg.norm(aa, axis=0)
project = np.dot(ss, aa) / norm_aa + gms
project_mean = np.mean(project, axis=1)
# 估计值投影图表
pl.scatter(project_mean, np.sum(u * norm_a, axis=1))
pl.title('est s project mean-number of right questions')
pl.xlabel('normal s-number')
pl.ylabel('right questions')
pl.show()
```

```python
norm_a = np.linalg.norm(a, axis=0)
real_project = np.dot(s, a) / norm_a + gm_s
real_project_mean = np.mean(real_project, axis=1)
# 真值投影图表
pl.scatter(real_project_mean, np.sum(u * norm_a, axis=1))
pl.title('real s project mean-number of right questions')
pl.xlabel('normal s-number')
pl.ylabel('right questions')
pl.show()
```
